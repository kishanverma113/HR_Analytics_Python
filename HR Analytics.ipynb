# MATPLOTLIB :-

```python
import matplotlib.pyplot as plt
```

```python
# Age Distribution
plt.figure(figsize=(8,5))
plt.hist(df["Age"],bins=15)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Employees")
plt.show()
```

```python
# Salary Distribution
plt.figure(figsize=(8,5))
plt.hist(df["Salary"],bins=20)
plt.title("Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Employees")
plt.show()
```

```python
# Gender Count
df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
# Department Count
df["Department"].value_counts().plot(kind="bar")
plt.title("Department Wise Employees")
plt.xticks(rotation=45)
plt.show()
```

```python
# Attrition Count
df["Attrition"].value_counts().plot(kind="bar")
plt.title("Attrition Analysis")
plt.show()
```

```python
# Job Satisfaction
df["JobSatisfaction"].value_counts().plot(kind="bar")
plt.title("Job Satisfaction")
plt.show()
```

```python
# Work Life Balance
df["WorkLifeBalance"].value_counts().plot(kind="bar")
plt.title("Work Life Balance")
plt.show()
```

```python
# Scatter Plot
plt.scatter(df["Age"],df["Salary"])
plt.title("Age vs Salary")
plt.xlabel("Age")
plt.ylabel("Salary")
plt.show()
```

```python
# Line Plot
plt.plot(df["YearsAtCompany"],df["Salary"])
plt.title("Years at Company vs Salary")
plt.xlabel("Years")
plt.ylabel("Salary")
plt.show()
```

```python
# Pie Chart
df["Attrition"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.title("Attrition Percentage")
plt.ylabel("")
plt.show()
```

```python
# Box Plot
plt.boxplot(df["Salary"])
plt.title("Salary Box Plot")
plt.show()
```

```python
# Overtime Analysis
df["OverTime"].value_counts().plot(kind="bar")
plt.title("Overtime Employees")
plt.show()
```

## SEABORN :-

```python
import seaborn as sns
```

```python
# Countplot
plt.figure(figsize=(6,4))
sns.countplot(data=df,x="Attrition")
plt.show()
```

```python
# Gender Wise Attrition
plt.figure(figsize=(6,4))
sns.countplot(data=df,x="Gender",hue="Attrition")
plt.show()
```

```python
# Salary Distribution
plt.figure(figsize=(7,5))
sns.histplot(df["Salary"],bins=20,kde=True)
plt.show()
```

```python
# Age Distribution
plt.figure(figsize=(7,5))
sns.histplot(df["Age"],bins=15,kde=True)
plt.show()
```

```python
# Boxplot Salary
plt.figure(figsize=(6,5))
sns.boxplot(y=df["Salary"])
plt.show()
```

```python
# Boxplot Age by Attrition
plt.figure(figsize=(7,5))
sns.boxplot(data=df,x="Attrition",y="Age")
plt.show()
```

```python
# Violin Plot
plt.figure(figsize=(7,5))
sns.violinplot(data=df,x="Attrition",y="Salary")
plt.show()
```

```python
# Scatter Plot
plt.figure(figsize=(7,5))
sns.scatterplot(data=df,x="Age",y="Salary",hue="Attrition")
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(16,10))
sns.heatmap(
    df.corr(numeric_only=True),
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()
```
